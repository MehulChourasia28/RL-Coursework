```python
# ineffiecient
q_vals = np.array([[6,2,3],
                  [4,5,6],
                  [7,1,9]])

action_space = np.array([[True, True, False],
                         [False, False, False],
                         [True, True, False]])

valid_actions = np.where(action_space==True)   # arr of rows and arr of cols

valid_actions = list(zip(*valid_actions))  # turn into tuple pairs

value_idx_dict = {q_vals[idx]:idx for idx in valid_actions}
print(value_idx_dict)

best_action_value, best_action_idx = max(value_idx_dict.items())
print(best_action_value, best_action_idx)
```

```python
# efficient
q_vals = np.array([[5,2,3],
                  [4,5,6],
                  [7,1,9]])

action_space = np.array([[True, True, False],
                         [False, False, False],
                         [True, True, False]])

# np.where(condition, val if true, val if false)
valid_actions = np.where(action_space, q_vals, -np.inf) # invalid actions masked
print(valid_actions)

# exploit
flat_idx = np.argmax(valid_actions)
print(flat_idx) 

best_idx = np.unravel_index(flat_idx, q_vals.shape)
print(best_idx)

# explore
flat_valid_idxs = np.flatnonzero(action_space)
print(flat_valid_idxs)

idx = np.random.choice(flat_valid_idxs)

random_idx = np.unravel_index(idx, q_vals.shape)
print(random_idx)
```

```python
win_lines = [np.array([0,1,2]), # top row
                          [3,4,5], # mid row
                          [6,7,8], # bot row
                          [0,3,6], # 1st col
                          [1,4,7], # 2nd col
                          [2,5,8], # 3rd col
                          [0,4,8], # main diag
                          [2,4,6]  # other diag
                          ]

board = np.array([1,1,1,0,0,0,2,2,2])
board1 = (board==1)

is_win = False
for line in win_lines:

    if board1[line].all():   # if all elements in arr are True, then True
        is_win = True
        break
print(is_win)
```
